```python
# 1. Question
# Q1: Identify missing or incorrect data in the dataset and apply appropriate preprocessing steps to clean it
```

```python
import pandas as pd
file_path = 'dataset.csv'
data = pd.read_csv(file_path)
```

```python
print("Original Dataset:")
print(data.head())
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
for column in ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']:
    mean_value = data[column].mean()
    data[column] = data[column].fillna(mean_value)
```

```python
data['sex'] = data['sex'].replace('unknown', pd.NA)
data = data.dropna(subset=['sex']) 
```

```python
outlier_condition = (data['body_mass_g'] > 6000)  # Assuming 6000g is an outlier threshold
data.loc[outlier_condition, 'body_mass_g'] = data['body_mass_g'].mean()  # Replace outliers with mean
```

```python
print("\nCleaned Dataset:")
print(data.head())
```

```python
cleaned_file_path = 'cleaned_penguin_data.csv'
data.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to {cleaned_file_path}")
```

```python
# 2.Question
#Q2: What is the average body_mass_g for Gentoo penguins?
```

```python
# Filter the dataset for Gentoo penguins
gentoo_penguins = data[data['species'] == 'Gentoo']

# Calculate the average body mass for Gentoo penguins, excluding any missing values
average_body_mass = gentoo_penguins['body_mass_g'].mean()

# Display the result
print(f"\nAverage body_mass_g for Gentoo penguins: {average_body_mass:.2f} g")
```

```python
# 3.Question
#Q3: How do the distributions of bill_length_mm and bill_depth_mm differ between the three penguin species? 
# Analyze the skewness and kurtosis of each feature for different species. (code and explanation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='species', y='bill_length_mm', data=data)
plt.title('Distribution of Bill Length (mm) by Species')
plt.ylabel('Bill Length (mm)')

# Plotting bill_depth_mm distribution for each species
plt.subplot(1, 2, 2)
sns.boxplot(x='species', y='bill_depth_mm', data=data)
plt.title('Distribution of Bill Depth (mm) by Species')
plt.ylabel('Bill Depth (mm)')

plt.tight_layout()
plt.show()
```

```python
# Calculate skewness and kurtosis for bill_length_mm and bill_depth_mm by species
stats_summary = data.groupby('species').agg(
    bill_length_skew=('bill_length_mm', skew),
    bill_length_kurtosis=('bill_length_mm', kurtosis),
    bill_depth_skew=('bill_depth_mm', skew),
    bill_depth_kurtosis=('bill_depth_mm', kurtosis)
).reset_index()

# Display the summary statistics
print("Skewness and Kurtosis Summary:")
print(stats_summary)
```

```python
# 4.Question
#Q4: Identify which features in the dataset have outliers. Provide the method used to detect them and visualize the outliers.
```

```python
# Function to identify outliers using IQR
def identify_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python

# Identify outliers for each feature of interest
outliers_bill_length = identify_outliers_iqr(data, 'bill_length_mm')
outliers_bill_depth = identify_outliers_iqr(data, 'bill_depth_mm')
outliers_flipper_length = identify_outliers_iqr(data, 'flipper_length_mm')
outliers_body_mass = identify_outliers_iqr(data, 'body_mass_g')

# Combine all outliers into a single DataFrame for visualization purposes
outliers_combined = pd.concat([outliers_bill_length, outliers_bill_depth, outliers_flipper_length, outliers_body_mass]).drop_duplicates()
```

```python
# Visualization of outliers using boxplots
plt.figure(figsize=(14, 10))

# Boxplot for bill_length_mm
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='bill_length_mm', data=data)
plt.title('Bill Length (mm) by Species')

# Boxplot for bill_depth_mm
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='bill_depth_mm', data=data)
plt.title('Bill Depth (mm) by Species')

# Boxplot for flipper_length_mm
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='flipper_length_mm', data=data)
plt.title('Flipper Length (mm) by Species')

# Boxplot for body_mass_g
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='body_mass_g', data=data)
plt.title('Body Mass (g) by Species')

plt.tight_layout()
plt.show()
```

```python
# Display the identified outliers
print("\nIdentified Outliers:")
print(outliers_combined[['species', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']])
```

```python
# 5.Question
#Q5: Does this dataset contribute to the curse of dimensionality? If yes perform PCA.
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Select features for PCA
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
X = data[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['species'] = data['species'].reset_index(drop=True)
```

```python
# Visualize PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='species', data=pca_df, palette='Set1')
plt.title('PCA of Penguin Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.grid()
plt.show()
```

```python
# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by each principal component: {explained_variance}")
```

```python
# 7.Question
# Q7: Find the maximum flipper_length_mm for each combination of species and island. Which species has the longest flippers on each island?
```

```python
# Group by species and island, then find the maximum flipper length
max_flipper_length = data.groupby(['species', 'island'])['flipper_length_mm'].max().reset_index()

# Determine which species has the longest flippers on each island
longest_flippers = max_flipper_length.loc[max_flipper_length.groupby('island')['flipper_length_mm'].idxmax()]

# Display the results
print("Maximum Flipper Lengths by Species and Island:")
print(max_flipper_length)

print("\nSpecies with the Longest Flippers on Each Island:")
print(longest_flippers)
```
